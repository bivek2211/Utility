```
import spacy

# load large english NLP model
nlp = spacy.load("en_core_web_sm")           # load model package "en_core_web_sm"

text = """London is the capital and most populous city of England and 
the United Kingdom.  Standing on the River Thames in the south east 
of the island of Great Britain, London has been a major settlement 
for two millennia. It was founded by the Romans, who named it Londinium.
"""

# parsing the text with spaCy
doc = nlp(text)


# doc now contains a parsed version of text. We can use it to do anything e want !
# let's print the name of entity that were detected
for entity in doc.ents:
  print(f"{entity.text} ({entity.label_})")
```

```
## Tokenization

# i=0
# for token in doc:
#   print(f"{i} --> {token.text}")
#   i=i+1
  
print([w.text for w in doc])
```

```
# textacy implements common data extraction algorithms
# implementing semi-structured statement extraction
# (based on london is/be)

import textacy.extract

statements = textacy.extract.semistructured_statements(doc,"London")

print("Things about London")

for statement in statements:
  subject, verb, fact = statement
  print(f"  fact : {fact}")
  print(f"  verb : {verb}")
  print(f"  subject : {subject}")
  print("------------------------------------------------------------------")
```

```
# Extract noun chunks that appear
noun_chunks = textacy.extract.noun_chunks(doc, min_freq=1)

# Convert noun chunks to lowercase strings
noun_chunks = map(str, noun_chunks)
noun_chunks = map(str.lower, noun_chunks)

# Print out any nouns that are at least 2 words long
for noun_chunk in set(noun_chunks):
    if len(noun_chunk.split(" ")) > 1:
        print(noun_chunk)
```
