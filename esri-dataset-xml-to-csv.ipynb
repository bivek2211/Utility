```python
import glob
import pandas as pd
import xml.etree.ElementTree as ET

TRAINING_IMAGES_DIR = "../input/training_data/training_data/"
TEST_IMAGES_DIR = "../input/test_data_images/test_data_images/"

MIN_NUM_IMAGES_REQUIRED_FOR_TRAINING = 10
MIN_NUM_IMAGES_SUGGESTED_FOR_TRAINING = 100

MIN_NUM_IMAGES_REQUIRED_FOR_TESTING = 3

# # output .csv file names/locations
# TRAINING_DATA_DIR = os.getcwd() + "/" + "training_data"
# TRAIN_CSV_FILE_LOC = TRAINING_DATA_DIR + "/" + "train_labels.csv"
# EVAL_CSV_FILE_LOC = TRAINING_DATA_DIR + "/" + "eval_labels.csv"
```

```python
def xml_to_csv(path):
    xml_list = []
    for xml_file in glob.glob(path + 'labels/*.xml'):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            value = (root.find('filename').text, int(root.find('size')[0].text), int(root.find('size')[1].text), member[0].text,
                     float(member[1][0].text), float(member[1][1].text), float(member[1][2].text), float(member[1][3].text))
            xml_list.append(value)
        # end for
    # end for

    column_name = ['filename', 'width', 'height', 'name', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df

print("converting xml training data . . .")
trainCsvResults = xml_to_csv(TRAINING_IMAGES_DIR)
print("training xml to .csv conversion successful")
trainCsvResults
```

```python
trainCsvResults.to_csv('train_labels.csv',index=False)
```
